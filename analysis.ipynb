```python
%matplotlib inline

from short_vix_algo import ShortVixAlgo

#full period backtesting
setup_coef=0.6
enter_coef=0.95
ticker='TVIX'
initial_capital=10000

algo = ShortVixAlgo(setup_coef=setup_coef,
                    enter_coef=enter_coef,
                    ticker=ticker,
                    initial_capital=initial_capital)

data = algo.run_backtesting()
algo.print_algo_performance(data)

algo.plot_backtesting_results(data,['close'])
algo.plot_backtesting_results(data,['capital'])
```

```python
# specific periods 
start = 20110101
end = 20130101
df1 = data[(data['date_int']>=start) & (data['date_int']<=end)]

algo.plot_backtesting_results(df1,['close'])
algo.plot_backtesting_results(df1,['capital'])
```

```python
# specific periods 
start = 20160101
end = 20170101
df1 = data[(data['date_int']>=start) & (data['date_int']<=end)]

algo.plot_backtesting_results(df1,['close'])
algo.plot_backtesting_results(df1,['capital'])
```
